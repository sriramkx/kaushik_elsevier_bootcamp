```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://ticketnew.com/movies/chennai")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
soup
```

```python
all=soup.find_all("div",{"class":"DesktopRunningMovie_detail__uzstd"})
all
```

```python
len(all)
```

```python
all[0]
```

```python
all[0].find_all("div",{"class":"DesktopRunningMovie_movTitle__Q1pOY"})
```

```python
all[0].find_all("div",{"class":"DesktopRunningMovie_movTitle__Q1pOY"})[0].text
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("div",{"class":"DesktopRunningMovie_movTitle__Q1pOY"})[0].text)
    print(item.find_all("div",{"class":"DesktopRunningMovie_point__YtD_w"})[0].text)
    print('\n')
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["movie_name"]=item.find_all("div",{"class":"DesktopRunningMovie_movTitle__Q1pOY"})[0].text
    #print(item.find_all("div",{"class":"tn-inline"})[0].text)
    d["movie_category"]=item.find_all("div",{"class":"DesktopRunningMovie_point__YtD_w"})[0].text
    
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```
