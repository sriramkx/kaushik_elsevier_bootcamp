```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://property.sulekha.com/commercial-office-space-for-rent/guindy-chennai")
c = r.content
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
print(soup.prettify())
```

```python
all=soup.find_all("li",{"class":"list-item"})
all
```

```python
len(all)
```

```python
all[0]
```

```python
all[0].find_all("h3") #price is defined by using tag and dimension array value
```

```python
all[0].find_all("h3")[0].text # to get only text
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("h3")[0].text)
```

```python
l=[]
for item in all:   #to display all price value in the website
    d={}
    d["Sqft area"]=item.find_all("h3")[0].text.replace("Office Space for Rent in Guindy","")
    d["Location"]=item.find_all("span",{"class":"location f-icon"})[0].text.replace(" ","")
    d["month_rent"]=item.find_all("i",{"class":"f-icon"})[0].text.replace(" ",""),"/month"
    d["deposit"]=item.find_all("span",{"class":"deposit"})[0].text.replace("Deposit:","").replace(" ","")
    try:
        d["building name"]=item.find_all("span",{"class":"buztitle"})[0].text.replace(" ","")
    except:
        d["building name"]=None
    
    
    #print(item.find_all("span",{"class":"feature primary"})[0].text.replace(" ","") )
    #print(item.find_all("div",{"class":"locality"})[0].text.replace(" ","").replace("ViewMap","") )
    d["broker name"]=item.find_all("span",{"class":"bsnsname"})[0].text.replace(" ","").replace("Broker:","") 
    d["posted on"]=item.find_all("span",{"class":"postdate f-icon"})[0].text.replace(" ","") 
    '''try:
         d["requirements"]=item.find_all("div",{"class":"spec icon primary"})[0].text.replace(" ","")
    except:
         d["requirements"]=None
         '''
   
    l.append(d)
```
